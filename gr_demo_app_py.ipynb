```python
# Step 2: Import libraries
import gradio as gr
import os
```

```python

import os
os.environ["HF_API_KEY"] = HF_API_KEY

from huggingface_hub import InferenceClient
client = InferenceClient(token=os.environ["HF_API_KEY"])


def chat_with_model(prompt):
    try:
        messages = [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user",   "content": prompt}
        ]

        response = client.chat_completion(
            model="meta-llama/Llama-3.1-8B-Instruct",
            messages=messages,
            max_tokens=500
        )

        print(response.choices[0].message.content)
        print(response.usage.prompt_tokens)

        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
# Step 5: Create a Gradio interface
iface = gr.Interface(
    fn=chat_with_model,
    inputs=gr.Textbox(lines=2, placeholder="Ask me anything..."),
    outputs="text",
    title="Chat with AI Models",
    description="Ask the model any questions."
)
```

```python
# Step 6: Launch the interface
iface.launch()
```

ADDING MEMORY TO THE MODEL

```python
from huggingface_hub import InferenceClient

client = InferenceClient(token=os.environ["HF_API_KEY"])

chat_history = []

def chat_with_model(prompt):
    try:
        # Add the user’s prompt to history
        chat_history.append({"role": "user", "content": prompt})

        # Call the model with the full history
        response = client.chat_completion(
            model="meta-llama/Llama-3.1-8B-Instruct",
            messages=chat_history,
            max_tokens=150,
            temperature=0.7
        )

        reply = response.choices[0].message.content

        chat_history.append({"role": "assistant", "content": reply})

        return reply

    except Exception as e:
        return f"Error: {str(e)}"
```

```python
chat_with_model("Whst is my first question?")
```

```python
import gradio as gr
iface = gr.Interface(
    fn=chat_with_model,
    inputs=gr.Textbox(lines=2, placeholder="Ask me anything..."),
    outputs="text",
    title="Chat with AI Models",
    description="Ask the model any questions."
)
```

```python
iface.launch()
```

CHANGING INTERFACE FROM A SIMPLE INPUT AND OUTPUT INTERFACE TO A CHAT INTERFACE

```python
from huggingface_hub import InferenceClient

client = InferenceClient(token=os.environ["HF_API_KEY"])

chat_history = []

def chat_with_model(prompt, history):
    try:
        # Add the user’s prompt to history
        chat_history.append({"role": "user", "content": prompt})

        # Call the model with the full history
        response = client.chat_completion(
            model="meta-llama/Llama-3.1-8B-Instruct",
            messages=chat_history,
            max_tokens=150,
            temperature=0.7
        )

        reply = response.choices[0].message.content

        chat_history.append({"role": "assistant", "content": reply})

        return reply

    except Exception as e:
        return f"Error: {str(e)}"
```

```python
import gradio as gr
iface = gr.ChatInterface(
    fn=chat_with_model,
    type='messages',
    title="Chat with AI Models",
    description="Ask the model any questions."
)
```

```python
iface.launch()
```
