```python
# Step 2: Import libraries
import gradio as gr
import os
```

```python
from google.colab import userdata
from huggingface_hub import InferenceClient

# Initialize everything
token = userdata.get('hf-token')
client = InferenceClient(token=token)

def chat_with_model(prompt):
    try:
        messages = [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ]

        # Now 'client' is defined, so no more yellow lines!
        response = client.chat_completion(
            model="meta-llama/Llama-3.2-3B-Instruct", # Note: Using a common accessible model
            messages=messages,
            max_tokens=500
        )

        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}"
# Test it
print(chat_with_model("Explain AI in one sentence."))
```

```python
# Step 5: Create a Gradio interface
iface = gr.Interface(
    fn=chat_with_model,
    inputs=gr.Textbox(lines=2, placeholder="Ask me anything..."),
    outputs="text",
    title="Chat with AI Models",
    description="Ask the model any questions."
)
```

```python
# Step 6: Launch the interface
iface.launch()
```

```python
from huggingface_hub import InferenceClient

# Initialize everything
token = userdata.get('hf-token')
client = InferenceClient(token=token)

chat_history = []

def chat_with_model(prompt):
    try:
        # Add the userâ€™s prompt to history
        chat_history.append({"role": "user", "content": prompt})

        # Call the model with the full history
        response = client.chat_completion(
            model="meta-llama/Llama-3.1-8B-Instruct",
            messages=chat_history,
            max_tokens=150,
            temperature=0.7
        )

        reply = response.choices[0].message.content

        chat_history.append({"role": "assistant", "content": reply})

        return reply

    except Exception as e:
        return f"Error: {str(e)}"
```

```python
chat_with_model("Whst is my best color ")
```

```python
import gradio as gr
iface = gr.Interface(
    fn=chat_with_model,
    inputs=gr.Textbox(lines=2, placeholder="Ask me anything..."),
    outputs="text",
    title="Chat with AI Models",
    description="Ask the model any questions."
)
```

```python
iface.launch()
```
